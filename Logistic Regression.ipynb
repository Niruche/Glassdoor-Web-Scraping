```python
df = pd.read_csv('glassdoor_full_dataset.csv')
df['date'] = pd.to_datetime(df.date)
```

```python
df.head()
```

```python
df['subject'] = df.subject.fillna('')
df['pros'] = df.pros.fillna('')
df['cons'] = df.cons.fillna('')

df['summary'] = df.subject + ' ' + df.pros + ' ' + df.cons
```

```python
df['subject'] = df.apply(lambda row: nltk.word_tokenize(row['subject']), axis=1)
df['pros'] = df.apply(lambda row: nltk.word_tokenize(row['pros']), axis=1)
df['cons'] = df.apply(lambda row: nltk.word_tokenize(row['cons']), axis=1)


lemmatizer = WordNetLemmatizer()

df['subject'] = df['subject'].apply(lambda x: ' '.join([lemmatizer.lemmatize(w) for w in x] ))
df['pros'] = df['pros'].apply(lambda x: ' '.join([lemmatizer.lemmatize(w) for w in x] ))
df['cons'] = df['cons'].apply(lambda x: ' '.join([lemmatizer.lemmatize(w) for w in x] ))
```

```python
# only keep alphabet
df['summary'] = df['summary'].map(lambda x: re.sub('[^a-zA-Z]', ' ', x))
df['summary'] =df['summary'].map(lambda x: re.sub(r'\s+', ' ', x))
```
