## Scrapping

```python
stem_template_url = 'https://www.glassdoor.co.uk/Reviews/Pret-A-Manger-Reviews-E13199_P'
end_template_url = '.htm'

# define an empty conrtainer
output_list = []

wait = WebDriverWait(driver, 10)

for i in range(1, 71):
    url = stem_template_url + str(i) + end_template_url
    #print(url)
    #go to webpage
    #url = 'https://www.glassdoor.co.uk/Reviews/Pret-A-Manger-Reviews-E13199.htm'
    driver.get(url)
    
    #cookies_accept = driver.find_element_by_xpath("//*[@id='_evidon-accept-button']")
    #cookies_accept.click()
    
    # Sleep to allow page to load
    sleep(2)

    review_container = driver.find_elements_by_class_name('hreview')

    
    for review in review_container:
        
        glassdoor_dict = {}
        
        date = review.find_element_by_class_name('date').text
        author = review.find_element_by_class_name('author').text
        star = review.find_element_by_class_name('value-title').get_attribute("title")
        subject = review.find_element_by_class_name('summary').text
        tenure = review.find_element_by_class_name('mainText').text
        comments = review.find_element_by_class_name('col-sm-11').text
                                
        #description = review.find_element_by_class_name('description').text
        try:
            overview = review.find_element_by_class_name('reviewBodyCell').text
        except:
            overview = 'No Data'
        try:
            advice = review.find_element_by_class_name(' adviceMgmt').text
        except:
            advice = 'Not available'
            
        glassdoor_dict['date'] = date 
        glassdoor_dict['author'] = author
        glassdoor_dict['star'] = star
        glassdoor_dict['subject'] = subject
        glassdoor_dict['tenure'] = tenure
        glassdoor_dict['comments'] = comments
        glassdoor_dict['overview'] = overview
       
    
        # Look at one
        
        try:
        
            section_rating = review.find_element_by_class_name("stars__StarsStyles__gdStars")

            rating_info = section_rating.find_element_by_class_name("stars__StarsStyles__subRatings").find_element_by_class_name("undecorated")

            elem_ratings = rating_info.find_elements_by_tag_name('li')

            for elem in elem_ratings:

                rat_name = elem.find_element_by_class_name('minor').get_attribute('innerText')
                rat_value= elem.find_element_by_class_name("gdRatings").get_attribute('title')

                glassdoor_dict[rat_name] = rat_value
                
        except:
            
            pass
        


        
        
        output_list.append(glassdoor_dict)

```

```python
df2['overview_split'] = df2['comments'].str.split('\n')
df2['pros'] = df2.comments.str.extract('Pros\n(?P<Pros>.*)')
df2['cons'] = df2.comments.str.extract('Cons\n(?P<Pros>.*)')
df2['advice'] = df2.comments.str.extract('Advice to Management\n(?P<Pros>.*)')
df2['location'] = df2.author.str.split('in').str[1].str.split(',').str[0]
df2['member_type'] = df2.author.str.split('-').str[-1].str.split('in').str[0]
df2['employee_status'] = df2['author'].str.split(' - ').str[0]
df2['subject'] = df2.subject.str.replace('"', '')
```

```python
def overview_grader(list_of_opinions):
    
    recommended = 'No Data'
    outlook = 'No Data'
    CEO = 'No Data'
    
    # Loop through for recommended
    for word in list_of_opinions:
        
        if word == 'Recommends':
            recommended = True
            break
        elif word == "Doesn't Recommend":
            recommended = False
            break
        else:
            pass
        
    # Loop through for outlook
    for word in list_of_opinions:
        
        if word == 'Positive Outlook':
            outlook = 'positive'
            break
        elif word == "Negative Outlook":
            outlook = 'negative'
            break
        elif word == "Neutral Outlook":
            outlook = 'neutral'
            break
        else:
            pass
        
     # Loop through for ceo
    for word in list_of_opinions:
        
        if word == 'Approves of CEO':
            CEO = 'positive'
            break
        elif word == "Disapproves of CEO":
            CEO = 'negative'
            break
        elif word == "No Opinion of CEO":
            CEO = 'neutral'
            break
        else:
            pass
    
    return recommended, outlook, CEO
```

```python
df3 = df2.copy()
```

```python
df3['recommend'], df3['outlook'], df3['ceo'] =  zip(*df3['overview_split'].map(overview_grader))
df3.head()
```

## Review overall rating column

```python
clean_table['star'] = clean_table['star'].astype(str).str[0]
clean_table['star'] = clean_table['star'].astype(int)


clean_table['Career Opportunities'] = clean_table['Career Opportunities'].fillna('0')
clean_table['Career Opportunities'] = clean_table['Career Opportunities'].astype(str).str[0]
clean_table['Career Opportunities'] = clean_table['Career Opportunities'].astype(int)

clean_table['Compensation and Benefits'] = clean_table['Compensation and Benefits'].fillna('0')
clean_table['Compensation and Benefits'] = clean_table['Compensation and Benefits'].astype(str).str[0]
clean_table['Compensation and Benefits'] = clean_table['Compensation and Benefits'].astype(int)

clean_table['Culture & Values'] = clean_table['Culture & Values'].fillna('0')
clean_table['Culture & Values'] = clean_table['Culture & Values'].astype(str).str[0]
clean_table['Culture & Values'] = clean_table['Culture & Values'].astype(int)

clean_table['Senior Management'] = clean_table['Senior Management'].fillna('0')
clean_table['Senior Management'] = clean_table['Senior Management'].astype(str).str[0]
clean_table['Senior Management'] = clean_table['Senior Management'].astype(int)

clean_table['Work/Life Balance'] = clean_table['Work/Life Balance'].fillna('0')
clean_table['Work/Life Balance'] = clean_table['Work/Life Balance'].astype(str).str[0]
clean_table['Work/Life Balance'] = clean_table['Work/Life Balance'].astype(int)

```

```python
def star_grouper(star):

    if star == 0:
        return "NA"
    elif star >= 4:
        return "Positive"
    elif star <= 2:
        return "Negative"
    else:
        return 'Neutral'
```

```python
clean_table['star_group'] = clean_table['star'].apply(star_grouper)

clean_table['Career Opportunities'] = clean_table['Career Opportunities'].apply(star_grouper)
clean_table['Compensation and Benefits'] = clean_table['Compensation and Benefits'].apply(star_grouper)
clean_table['Culture & Values'] = clean_table['Culture & Values'].apply(star_grouper)
clean_table['Senior Management'] = clean_table['Senior Management'].apply(star_grouper)
clean_table['Work/Life Balance'] = clean_table['Work/Life Balance'].apply(star_grouper)

```
